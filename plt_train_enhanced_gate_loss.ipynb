```python
filea = [i for i in open("/root/autodl-tmp/Qwen2-VL-Finetune/csvqa_mamoev02.log", "r").readlines() if i.startswith("{'loss': ")]
fileb = [i for i in open("/root/autodl-tmp/Qwen2-VL-Finetune/chartqa_kradapter_v0.log", "r").readlines() if i.startswith("{'loss': ")]
```

```python
import ast

import matplotlib.pyplot as plt

def parse_losses(lines):
    epochs, losses = [], []
    for s in lines:
        d = ast.literal_eval(s.strip())
        epochs.append(d.get('epoch'))
        losses.append(d.get('loss'))
    return epochs, losses

ea, la = parse_losses(filea)
eb, lb = parse_losses(fileb)

plt.figure(figsize=(8,5))
plt.plot(ea, la, '-o', label='filea')
plt.plot(eb, lb, '-o', label='fileb')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Loss vs Epoch')
plt.legend()
plt.grid(True)
plt.show()
```
